```python
import random

board = [[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
all_moves = [(x, y) for x in range(3) for y in range(3)]
USER_SYMBOL = 'X'
COMPUTER_SYMBOL = 'O'

def main():
  winner = None
  while True:
    x, y = None, None
    valid_move = False
    while (x is None or y is None) and not valid_move:
      x, y = get_user_move()
      valid_move = is_valid_move(x, y)
    mark_move(x, y, USER_SYMBOL)
    if get_winner() == USER_SYMBOL:
      print("You won!")
      print_board()
      break
    x, y = get_computer_move()
    if x is None or y is None:
      print("No more moves possible!")
      print_board()
      break
    mark_move(x, y, COMPUTER_SYMBOL)
    if get_winner() == COMPUTER_SYMBOL:
      print("Computer won!")
      print_board()
      break

def mark_move(x, y, symbol):
  all_moves.remove((x, y))
  board[x][y] = symbol

def get_computer_move():
  if len(all_moves) == 0:
    return None, None
  return random.choice(all_moves)

def get_user_move():
  user_input = input("Your move in \"x, y\" format")
  if ',' not in user_input:
    print("Input in wrong format")
    return None, None
  x_str, y_str = user_input.split(',')
  try:
    x = int(x_str.strip())
    y = int(y_str.strip())
  except ValueError:
    print("Input in wrong format")
    return None, None
  return x, y

def is_valid_move(x, y):
  if x is None or y is None:
    print("Input in wrong format")
    return False
  if x < 0 or x > 2 or y < 0 or y > 2:
    print("Invalid move")
    return False
  return board[x][y] != ' '

def get_winner():
  for i in range(3):
    if board[i][0] != ' ' and board[i][0] == board[i][1] and board[i][0] == board[i][2]:
      return board[i][0]
    if board[0][i] != ' ' and board[0][i] == board[1][i] and board[0][i] == board[2][i]:
      return board[0][i]
  if board[0][0] != ' ' and board[0][0] == board[1][1] and board[0][0] == board[2][2]:
    return board[0][0]
  if board[0][2] != ' ' and board[0][2] == board[1][1] and board[0][2] == board[0][2]:
    return board[0][2]
  return None

def print_board():
  print("Current board:")
  for x in range(3):
    print(''.join(board[x]))
```
